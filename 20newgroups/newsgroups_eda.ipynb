# Imports

```python
from collections import Counter
from sklearn.datasets import fetch_20newsgroups
import matplotlib.pyplot as plt
```

# Load dataset

```python
# Set the categories you want to include
cats = ['alt.atheism',
        'comp.graphics',
        'comp.os.ms-windows.misc',
        'comp.sys.ibm.pc.hardware',
        'comp.sys.mac.hardware',
        'comp.windows.x',
        'misc.forsale',
        'rec.autos',
        'rec.motorcycles',
        'rec.sport.baseball',
        'rec.sport.hockey',
        'sci.crypt',
        'sci.electronics',
        'sci.med',
        'sci.space',
        'soc.religion.christian',
        'talk.politics.guns',
        'talk.politics.mideast',
        'talk.politics.misc',
        'talk.religion.misc']
```

```python
# Downloads data where the package is installed
data_raw = fetch_20newsgroups(subset='all', remove=('headers', 'footers', 'quotes'), categories=cats)
```

# Exploration

```python
print(f"Total documents: {len(data_raw.data)}")

# Print an example
print(data_raw.data[16])
```

```python
data_raw.target_names
```

```python
# Class Distribution
news_train = fetch_20newsgroups(subset='train')

category_counts = Counter(news_train.target)


plt.figure(figsize=(12, 5))
plt.bar(news_train.target_names, category_counts.values())
plt.xticks(rotation=45)
plt.xlabel("News Category")
plt.ylabel("Count")
plt.title("Class Distribution in 20 Newsgroups")
plt.show()
```

```python
# Average characters and words

total_chars = 0
total_words = 0
all_words = []

for text in data_raw.data:
    total_chars += len(text)
    words = text.split()
    total_words += len(words)
    all_words += words

vocab = set(all_words)

print(f'AVERAGE CHARS: {total_chars/len(data_raw.data)}')
print(f'AVERAGE WORDS: {total_words/len(data_raw.data)}')
print(f'TOTAL WORDS: {total_words}')
print(f'VOCABULARY SIZE: {len(vocab)}')
```
